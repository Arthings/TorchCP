```python
import torch
import numpy as np
epsilon = 0.1
# 设置随机种子
torch.manual_seed(0)
np.random.seed(0)

# 生成非线性数据
x = torch.linspace(-np.pi, np.pi, 100).reshape(-1, 1)
y_truth = 0.1*x**2 + torch.cos(x)
y = y_truth  + torch.randn(x.size()) * epsilon

dataset = TensorDataset(x, y)
```
